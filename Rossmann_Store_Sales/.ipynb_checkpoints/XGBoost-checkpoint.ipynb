## 调整权重

```python
Model = joblib.load('Model_Parameter2/XGBoostReg_reduce_log2.pkl')
# from help import submission_log
# submission_log(Model, Save_File=True)

from help import rmspe_log
import datetime
starttime = datetime.datetime.now()
print('Calculating test score...')
sale_predict = Model.predict(X_test_reduce)
print('  Test score',rmspe_log( sale_predict, y_test_log))
endtime = datetime.datetime.now()
print('  Done!  \n  Using time:',(endtime - starttime).seconds,'sec\n')
```

```python
df_test = pd.read_csv('data/X_process.csv', index_col = 0)[:6*7*1115]
df_test['Sales'] = np.expm1(y_test_log)
df_test['Predicition'] = np.expm1(sale_predict)
df_test['Ratio']=df_test['Predicition']/df_test['Sales']
df_test['Error']=abs(df_test['Ratio']-1)
df_test['Weight']=df_test['Sales']/df_test['Predicition']
df_test.index.name = None
df_test.head()
```

```python
from help import rmspe
weights = [(0.990+(i/1000)) for i in range(20)]
Best_weights = []
for i in np.arange(1115):
    store_index = i+1
    sales = df_test[df_test['Store']==store_index]['Sales']
    predict = df_test[df_test['Store']==store_index]['Predicition']
    error = []
    for weight in weights:
        error.append( rmspe(predict*weight, sales) )
    Score=pd.Series(error,index=weights)
    Best_Score=Score[Score.values==Score.values.min()]
    Best_weights.append(Best_Score.index.values.tolist())
    
Best_weights = pd.Series([y for x in Best_weights for y in x], index=np.arange(1115)+1, dtype='float')
Best_weights.index.name = 'Store'
Best_weights.name = 'Adjust_Weight'
Best_weights.head()
```

```python
df_test = pd.merge(df_test,Best_weights,on='Store')
df_test['Adjust_Predicition'] = df_test['Predicition'] * df_test['Adjust_Weight']
df_test.head()
```

```python
from help import rmspe
rmspe(df_test['Adjust_Predicition'], df_test['Sales'])
```
